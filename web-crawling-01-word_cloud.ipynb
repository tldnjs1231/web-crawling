```python
for i in ',()-_!?\'"':
    butter = butter.replace(i, ' ')

words = butter.split()

dic = {}

for i in set(words):
    dic[i] = words.count(i)
```

```python
import turtle as t
from random import randint as ri

t.ht() # hideturtle()
t.pu() # penup()
t.speed(0)

t.colormode(255)
t.setup(900, 600)

for i in dic:
    t.color(ri(0, 255), ri(0, 255), ri(0, 255)) # pencolor()
    t.goto(ri(-400, 350), ri(-250, 200))
    t.write(i, font=('맑은고딕', (dic[i]+4)*3, 'bold'))

t.mainloop()
```
